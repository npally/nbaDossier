```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

kobe_df = pd.read_csv('kobe.csv').dropna()
harden_df = pd.read_csv('harden.csv').dropna()
kobePlayoff = pd.read_csv("kobePlayoff.csv").dropna()
hardenPlayoff = pd.read_csv("hardenPlayoff.csv")
```

```python
kobe_df["A/T"] = kobe_df["AST"] / kobe_df["TOV"]
harden_df["A/T"] = harden_df["AST"] / harden_df["TOV"]
```

```python
def get_data(stat, title):
    fig = plt.figure(figsize=(10,8))
    print("Career Averages")
    print("---------------")

    ax = fig.add_subplot(111)
    ax.set_title (f"{title} by Season")
    ax.plot(kobe_df.index,kobe_df[stat], 'y', label="Kobe Bryant", linewidth=3.0)
    ax.plot(harden_df.index, harden_df[stat], 'r', label="James Harden", linewidth=3.0)
    ax.legend()
    ax.grid(True)
    start, end = ax.get_xlim()
    ax.xaxis.set_ticks(np.arange(0, end, 1))
    harden_avg = round(harden_df[stat].mean(), 2)
    kobe_avg = round(kobe_df[stat].mean(),2)
    print(f'{title} Averages: Bryant {kobe_avg} | Harden {harden_avg}\n')

    print("First 11 Seasons")
    print("---------------")
    harden_avg = round(harden_df[stat].mean(), 2)
    kobe_avg = round(kobe_df[stat][:11].mean(),2)
    print(f'{title} Averages: Bryant {kobe_avg} | Harden {harden_avg}\n')

def get_pf_data(stat, title):
    fig = plt.figure(figsize=(10,8))
    print("Playoff Averages")
    print("---------------")

    ax = fig.add_subplot(111)
    ax.set_title (f"{title} by Season")
    ax.plot(kobePlayoff.index, kobePlayoff[stat], 'y', label="Kobe Bryant", linewidth=3.0)
    ax.plot(hardenPlayoff.index, hardenPlayoff[stat], 'r', label="James Harden", linewidth=3.0)
    ax.legend()
    ax.grid(True)
    start, end = ax.get_xlim()
    ax.xaxis.set_ticks(np.arange(0, end, 1))
    harden_avg = round(hardenPlayoff[stat].mean(), 2)
    kobe_avg = round(kobePlayoff[stat].mean(),2)
    print(f'{title} Averages: Bryant {kobe_avg} | Harden {harden_avg}\n')
```

### PER
What is PER?
Short for player efficiency rating, PER is a pace-adjusted rating of a player's per-minute productivity developed by John Hollinger (former VP of Basketball Operations for the Memphis Grizzlies). By pace-adjusted we mean the score is weighted the same for a high-pace team like the Bucks and slower-paced team like the Pistons. PER sums a player's positive accomplishments (made baskets, assists, steals, rebounds, etc) and subtracts negative accomplishments (missed baskets, turnovers, fouls). See more [here](https://www.espn.com/nba/columns/story?columnist=hollinger_john&id=2850240) and the formula [here](https://www.basketball-reference.com/about/per.html).

Harden's PER has been on a fairly linear trajectory since entering the league and has a higher PER score in 7 out of 11 seasons when compared with Bryant at the same point in their careers. Interestingly Bryant's lone MVP season (07-08, index 11) is not in his top 3 seasons of PER ratings. 

In terms of career PER Harden currently has the 10th highest in NBA history with Bryant ranked at 26th just one place behind Yao Ming. Currently, he top 3 players in league history are Michael Jordan, LeBron James, and Anthony Davis.

### Usage Rate

Usage Rate Formula

100*((Player’s Field Goal Attempts)+0.44*(Player’s Free Throw Attempts)+(Player’s Turnovers))*(Team’s Total Minutes)
/
((Team’s Total Field Goal Attempts)+0.44*(Team’s Total Free Throw Attempts)+Team’s Total Turnovers))*5*(Player’s Minutes)

A player's usage rate is essentially how many of a team's possessions end with that player whether by shot attempt, free throw, or turnover. Harden's usage really shoots up in his 4th year, which is the season he was traded to the Rockets. In of itself I wouldn't say usage % is super revealing but it can very useful when contextualizing other stats. A general rule of thumb is when a player's usage increases, his efficiency will decrease.

## True Shooting Percentage
Formula: 

TS% - True Shooting Percentage; the formula is PTS / (2 * TSA). True shooting percentage is a measure of shooting efficiency that takes into account field goals, 3-point field goals, and free throws.
TSA - True Shooting Attempts; the formula is FGA + 0.44 * FTA.

In my opinion TS% is a much better way to compare scoring efficiency. All forms of scoring are accounted for and here clearly show Harden to be the more efficient scorer.
